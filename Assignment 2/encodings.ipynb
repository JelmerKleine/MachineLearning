```python
import pandas as pd
```

```python
data = {'brand': ['ford', 'honda', 'bmw', 'ford', 'audi', 'bmw', 'ford'],
        'dist': [145340.0, 23458.0, 1487.0, 65459.0, 212984.0, 74678.0, 4327.0],
        'cond': ['good', 'average', 'good', 'bad', 'bad', 'average', 'good'],
        'serviced': ['T', 'F', 'F', 'T', 'F', 'U', 'T'],
        'price': [4499.0, 24600.0,45500.0, 12450.0, 2290.0, 20999.9, 27999.0],
        }
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.dtypes
```

```python
to_replace = { 'serviced': {'T': 1, 'F': 0, 'U':0} }
df = df.replace(to_replace)
```

```python
df['serviced'] = df['serviced'].astype(int)
```

```python
df
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordenc = OrdinalEncoder(categories=[['bad', 'average', 'good']])
```

```python
df['cond_ord'] = ordenc.fit_transform(df[['cond']])
```

```python
df
```

```python
# one-hot encoding
from sklearn.preprocessing import OneHotEncoder
ohenc = OneHotEncoder(sparse=False)
ohenc.fit_transform(df[['brand']])
```

```python
# One-hot encoding easy way
df = pd.get_dummies(df, columns=['brand'])
```
